```python
import pandas as pd
import folium
from folium import IFrame

# 엑셀 파일 불러오기
excel_file = '음식메뉴_추가.xlsx'
df = pd.read_excel(excel_file)

# 사용자에게 구 이름 입력 받기
target_gu = input("구 이름을 입력하세요: ")

# 입력한 구 이름에 해당하는 데이터 필터링
filtered_df = df[df['구 이름'] == target_gu]

# 별점을 기준으로 내림차순 정렬 후 상위 20개 선택
top_rated_df = filtered_df.sort_values(by='별점', ascending=False).head(20)

# 입력한 구의 중심 좌표 설정
center_lat, center_lon = top_rated_df['위도(y)'].mean(), top_rated_df['경도(x)'].mean()

# 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# 각 데이터행마다 지도에 마커 추가
for index, row in top_rated_df.iterrows():
    lat, lon = row['위도(y)'], row['경도(x)']
    popup_text = f"<div style='font-size: 12px;'>도로명주소: {row['도로명주소']}<br>카테고리: {row['카테고리']}<br>사이트: <a href='{row['사이트']}' target='_blank'>{row['사이트']}</a><br>별점: {row['별점']}</div>"
    popup = folium.Popup(IFrame(popup_text, width=300, height=100), parse_html=True)
    
    # 별점에 따라 마커 색상 설정
    if row['별점'] == 5.0:
        marker_color = 'red'
    elif row['별점'] >= 4.7:
        marker_color = 'orange'
    elif row['별점'] >= 4.5:
        marker_color = 'blue'
    else:
        marker_color = 'black'
    
    # 아이콘 생성
    marker_icon = folium.Icon(color=marker_color)
    marker = folium.Marker(location=[lat, lon], icon=marker_icon, popup=popup)
    marker.add_to(m)

# 지도를 창으로 띄우기
display(m)
```
